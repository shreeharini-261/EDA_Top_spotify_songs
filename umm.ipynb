removing $$ symbol from the dataset...ughhh

```python


print(df['Artists'].dtype)


df['Artists'] = df['Artists'].astype(str)
df['Title']=df['Title'].astype(str)


df['Artists'] = df['Artists'].apply(lambda x: x.replace('$', 's'))
df['Artists'] = df['Artists'].apply(lambda x: x.replace('*', 'g'))
df['Title'] = df['Title'].apply(lambda x: x.replace('$', 's'))
df['Title'] = df['Title'].apply(lambda x: x.replace('*', 'g'))
print(df)
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10, 8))
df_correlation = df.corr(numeric_only=True)

sns.heatmap(df_correlation, annot=True)
plt.show()
```

INTERACTIVE PLOTS!

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Danceability', bins=20, color='blue', label='Danceability')
sns.histplot(data=df, x='Energy', bins=20, color='orange', label='Energy')
plt.title('Distribution of Danceability and Energy')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

Relationship between Danceability and Energy...

```python
continent_counts = df['Continent'].value_counts()
plt.figure(figsize=(12, 8))
continent_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Songs by Continent')
plt.show()
```

```python
features = ['Danceability','Loudness','Speechiness','Acousticness','Energy','Valence','Instrumentalness']
df[features].hist(figsize=(10,8))
plt.title("Distribution Of Features" , fontsize=10)
plt.show()
```
